```python
#inputand output separation
ip = dataset.drop('Outcome' , axis = 1)
op = dataset['Outcome']
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(ip , op , test_size=0.2 , random_state = 4)
```

```python
#standardization of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
print(ip.shape ,x_train.shape , x_test.shape )
```

```python
#machine learning model
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf' ,C = 1000 , gamma = 0.01)
svc.fit(x_train , y_train)
```

```python
y_pred = svc.predict(x_test)
y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score
acc = accuracy_score(y_pred , y_test)
rec = recall_score(y_pred , y_test)
pre = precision_score(y_pred ,y_test)
f1 = f1_score(y_pred , y_test)

print('Accuracy:',acc)
print('recall:',rec)
print('precision:',pre)
print('f1 score:',f1)
```

```python
#making a predictive system

input_data = (6,148,72,35,0,33.6,0.627,50)
input_data_as_numpy = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy.reshape(1,-1)
standard_data = sc.transform(input_data_reshaped)
prediction = svc.predict(standard_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
